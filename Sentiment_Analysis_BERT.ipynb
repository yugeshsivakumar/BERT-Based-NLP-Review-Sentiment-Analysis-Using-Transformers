```python
import pandas as pd

# Sample data
data = {
    'review': [
        "This place is amazing! I loved the food and the ambiance.",
        "The service was terrible, and the food was not fresh.",
        "Pretty good overall, but the portions were small.",
        "Absolutely fantastic experience. Will come back for sure!",
        "Not worth the money. I expected much better quality.",
        "Great atmosphere, friendly staff, but the food could have been better.",
        "Disappointing experience. The wait time was too long, and the food was cold.",
        "A delightful meal! Everything was perfect, especially the dessert.",
        "The location is convenient, but the service needs improvement.",
        "One of the worst meals I've ever had. Will never return."
    ]
}

df = pd.DataFrame(data)

df.to_csv('sample_reviews.csv', index=False)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')


df = pd.read_csv('sample_reviews.csv')

def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt', truncation=True, max_length=512)
    result = model(tokens)
    return int(torch.argmax(result.logits)) + 1

df['sentiment'] = df['review'].apply(lambda x: sentiment_score(str(x)[:512]))

print(df)

df.to_csv('sentiment_analysis_output.csv', index=False)
```
